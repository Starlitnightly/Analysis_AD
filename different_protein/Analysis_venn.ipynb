```python
import pandas as pd
import sys
sys.path.append('C:/ProgramData/Anaconda3/Lib/site-packages')
from matplotlib_venn import venn2
```

```python
data1=pd.read_csv('SAMP8+SHED.csv')
data1.set_index(data1.columns[0],inplace=True)
data2=pd.read_csv('GSE121_result.csv')
data2.set_index(data2.columns[0],inplace=True)
```

```python
data1up=data1[data1['sig']=='up']
print(len(data1up))
data1down=data1[data1['sig']=='down']
print(len(data1down))
data2up=data2[data2['sig']=='up']
print(len(data2up))
data2down=data2[data2['sig']=='down']
print(len(data2down))
```

```python
ans1=list(set(data1up.index).intersection(set(data2up.index)))
ans2=list(set(data1down.index).intersection(set(data2down.index)))
print(ans1,ans2)
```

```python
venn2(subsets=[set(data1up.index),set(data2up.index)],set_labels=('SAMP8+SHED','SAMP8+GMS121'),set_colors=('r','green'))
```
