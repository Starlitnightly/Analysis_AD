```python
import ERgene
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
%matplotlib inline
from scipy import stats
```

```python
data=pd.read_csv('GSE_121.csv')
data.set_index(data.columns[0],inplace=True)
data.head(5)
```

```python
# Density

font1 = {
'weight' : 'bold',
'size'   : 18,
}
plt.figure(figsize=(20,15))
data.plot(kind = 'density', title = 'GSE101112_121')
plt.ylabel('Density',font1)
plt.title('CMS121',font1)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.tick_params(labelsize=15)
plt.savefig("GSE121density.png",dpi=300,bbox_inches = 'tight')
```

```python
#Mean expression of drug samples for each gene (row)
cms121 = data.loc[:, 'DD18_old121_rep1' : 'DD23_old121_rep6'].mean(axis = 1)
cms121.head()
```

```python
#Mean expression of conctrl samples for each gene (row)
ctrl = data.loc[:,'DD6_old_rep1':'DD11_old_rep6'].mean(axis = 1)
ctrl.head()
```

```python
fold = cms121/ctrl
log2fc=np.log2(fold)

plt.hist(log2fc)
plt.title("log2fc")
plt.savefig("GSE121fold.png",dpi=300,bbox_inches = 'tight')
```

```python
pvalue = []
for i in range(0, len(data)):
    ttest = stats.ttest_ind(data.iloc[i,6:12], data.iloc[i,0:6])
    pvalue.append(ttest[1])
pvalue[0:5]
```

```python
genearray = np.asarray(pvalue)
 
result = pd.DataFrame({'pvalue':genearray,'FoldChange':fold})
 
result['log(pvalue)'] = -np.log10(result['pvalue'])
result['log2FC'] = np.log2(result['FoldChange'])
result['sig'] = 'normal'
 
result['size']  =np.abs(result['FoldChange'])/10
 
result.loc[(result.log2FC> 0.25 )&(result.pvalue < 0.05),'sig'] = 'up'
result.loc[(result.log2FC< -0.25 )&(result.pvalue < 0.05),'sig'] = 'down'
result.to_csv('GSE121_result.csv')
```

```python
pp=plt.figure(figsize=(4,4))
ax = sns.scatterplot(x="log2FC", y="log(pvalue)",
                      hue='sig',
                      hue_order = ('up','down','normal'),
                      palette=("#E41A1C","green","grey"),
                     size='sig',sizes=(80, 83),
                     alpha=0.5,
                      data=result)
ax.set_ylabel('-log(pvalue)',font1)                                    
ax.set_xlabel('log2FC',font1)
plt.yticks(fontsize=15)
plt.xticks(fontsize=15)
plt.legend(fontsize=12,loc='best')
ax.tick_params(labelsize=15)
#ax.text(result.loc['Hook3']['log2FC'], result.loc['Hook3']['log(pvalue)'], 'Hook3',ha='center', va='bottom', fontdict=font1)
#ax.text(result.loc['Mif']['log2FC'], result.loc['Mif']['log(pvalue)'], 'Mif',ha='center', va='bottom', fontdict=font1)
#ax.text(result.loc['2410015M20Rik']['log2FC'], result.loc['2410015M20Rik']['log(pvalue)'], 'Mic13',ha='center', va='bottom', fontdict=font1)


plt.annotate('Hook3', xy=(result.loc['Hook3']['log2FC'], result.loc['Hook3']['log(pvalue)']), xytext=(1,2.5),
bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.5),
arrowprops=dict(arrowstyle='-',color='black'),size=15)

plt.annotate('Mif', xy=(result.loc['Mif']['log2FC'], result.loc['Mif']['log(pvalue)']), xytext=(-1,1),
bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.5),
arrowprops=dict(arrowstyle='-',color='black'),size=15)

plt.annotate('Mic13', xy=(result.loc['2410015M20Rik']['log2FC'], result.loc['2410015M20Rik']['log(pvalue)']), xytext=(-2,3),
bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.5),
arrowprops=dict(arrowstyle='-',color='black'),size=15)

plt.annotate('Rps9', xy=(result.loc['Rps9']['log2FC'], result.loc['Rps9']['log(pvalue)']), xytext=(-0.5,0.5),
bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.5),
arrowprops=dict(arrowstyle='-',color='black'),size=15)

ax.plot([result['log2FC'].min(),result['log2FC'].max()],[-np.log10(0.05),-np.log10(0.05)],linewidth=2, linestyle="--",color='black')
ax.plot([0.25,0.25],[result['log(pvalue)'].min(),result['log(pvalue)'].max()],linewidth=2, linestyle="--",color='black')
ax.plot([-0.25,-0.25],[result['log(pvalue)'].min(),result['log(pvalue)'].max()],linewidth=2, linestyle="--",color='black')

ax.tick_params(labelsize=12)
#plt.legend(bbox_to_anchor=(1.05,0), loc=3, borderaxespad=0)
plt.savefig("GSE121fire.png",dpi=300,bbox_inches = 'tight')
```

```python
fold_cutoff = 0.25
pvalue_cutoff = 0.05
 
filtered_ids = []
for i in range(0, len(result)):
    if (abs(-np.log2(fold[i])) >= fold_cutoff) and (pvalue[i] <= pvalue_cutoff):
        filtered_ids.append(i)
        
filtered = data.iloc[filtered_ids,:]
#filtered.to_csv('fi121GSE147.csv')

a=sns.clustermap(filtered, cmap='RdYlGn_r', standard_scale = 0,figsize=(3,9))
a.ax_heatmap.yaxis.set_tick_params(labelsize=12)
a.ax_heatmap.xaxis.set_tick_params(labelsize=12)
labels=a.ax_heatmap.xaxis.get_ticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
plt.savefig("GSE121clus.png",dpi=300,bbox_inches = 'tight')
```
